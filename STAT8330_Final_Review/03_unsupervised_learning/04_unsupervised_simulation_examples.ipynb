## What problem are we solving?

Unsupervised simulations show when clustering and dimension reduction recover structure and when they fail.

**Highest-probability exam pattern:** Vary cluster overlap, run an unsupervised method, and use known simulation labels only to evaluate recovery after the fact.

## Assumptions, intuition, and theory

In real unsupervised learning, labels are absent. Simulation labels are useful for studying failure modes. Overlap, scaling, non-spherical shapes, and irrelevant variables can make clusters harder to recover.

## Python method notes and documentation

`make_blobs` controls cluster overlap, `KMeans.fit_predict` returns clusters, `adjusted_rand_score` compares recovered clusters to known simulation labels, and `PCA.fit_transform` helps visualize structure.

## Fully self-contained worked example

Every code line below is commented so it is easy to scan under exam time pressure.

````python
import numpy as np  # Import numerical arrays and random-number tools.
import pandas as pd  # Import DataFrame tables for recovery summaries.
import matplotlib.pyplot as plt  # Import plotting tools for simulation examples.
from sklearn.cluster import KMeans  # Import k-means clustering.
from sklearn.datasets import make_blobs  # Import cluster simulator.
from sklearn.decomposition import PCA  # Import PCA for low-dimensional visualization.
from sklearn.metrics import adjusted_rand_score  # Import cluster recovery metric for simulated labels.
from sklearn.preprocessing import StandardScaler  # Import standardization before unsupervised methods.
RANDOM_SEED = 123  # Store the reproducibility seed.
np.random.seed(RANDOM_SEED)  # Fix legacy NumPy randomness.
plt.style.use('default')  # Use a predictable plotting style.
````

````python
def make_cluster_data(n=180, centers=3, cluster_std=0.8, random_state=123):  # Define a labeled cluster simulator for recovery checks.
    X, true_labels = make_blobs(  # Generate blob-shaped clusters with known labels.
        n_samples=n,  # Set the number of observations.
        centers=centers,  # Set the number of true clusters.
        cluster_std=cluster_std,  # Control cluster overlap.
        random_state=random_state,  # Fix the simulation seed.
    )  # Finish the simulator call.
    return X, true_labels  # Return data and true labels.
````

````python
rows = []  # Create a list for recovery results.
for cluster_std in [0.45, 0.90, 1.50]:  # Vary cluster overlap from easy to hard.
    X, true_labels = make_cluster_data(n=180, centers=3, cluster_std=cluster_std, random_state=RANDOM_SEED)  # Simulate clusters at this overlap level.
    X_scaled = StandardScaler().fit_transform(X)  # Standardize before k-means.
    labels = KMeans(n_clusters=3, n_init=20, random_state=RANDOM_SEED).fit_predict(X_scaled)  # Fit k-means and assign clusters.
    ari = adjusted_rand_score(true_labels, labels)  # Measure recovery against simulation truth.
    rows.append({'cluster_std': cluster_std, 'adjusted_rand_index': ari})  # Store recovery result.
results = pd.DataFrame(rows)  # Convert recovery rows to a DataFrame.
display(results)  # Display recovery under different overlaps.
X, true_labels = make_cluster_data(n=180, centers=3, cluster_std=1.5, random_state=RANDOM_SEED)  # Simulate the hardest example for visualization.
X_scaled = StandardScaler().fit_transform(X)  # Standardize the hardest data set.
pca_scores = PCA(n_components=2).fit_transform(X_scaled)  # Use PCA scores for a two-dimensional view.
labels = KMeans(n_clusters=3, n_init=20, random_state=RANDOM_SEED).fit_predict(X_scaled)  # Fit k-means to the hardest data set.
plt.figure(figsize=(6, 4))  # Create a visualization figure.
plt.scatter(pca_scores[:, 0], pca_scores[:, 1], c=labels, s=30, alpha=0.75)  # Plot PCA scores colored by recovered clusters.
plt.xlabel('PC1')  # Label the first PCA axis.
plt.ylabel('PC2')  # Label the second PCA axis.
plt.title('Unsupervised recovery gets harder as clusters overlap')  # Title the plot.
plt.show()  # Render the plot.
````
