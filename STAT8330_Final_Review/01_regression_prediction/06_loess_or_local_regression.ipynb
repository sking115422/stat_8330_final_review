
# LOESS or Local Regression

This notebook is intentionally self-contained. It does not import any local utility module, so you can open this file alone during study and see the full statistical workflow, simulation setup, model fitting, evaluation, plotting, and interpretation pattern in one place.



## What problem are we solving?

LOESS fits many small weighted regressions to estimate a smooth nonlinear mean curve without assuming a global polynomial form.

**Highest-probability exam pattern:** Use LOWESS to display a smooth curve and explain how span controls local smoothing.



## Assumptions, intuition, and theory

LOWESS assumes nearby x-values have similar mean responses. The span fraction controls how much of the data is used in each local fit. Small spans are flexible; large spans are smoother.



    ## Python method notes and documentation

    `lowess` returns sorted x and fitted y values, `frac` is the span, and `sample_weight` in `LinearRegression.fit` demonstrates the local-weighted-regression idea manually.

## Fully self-contained worked example

Every code line below is commented so the workflow can be scanned quickly under exam time pressure.

````python
import numpy as np  # Import numerical arrays and random-number tools.
import pandas as pd  # Import DataFrame tables for span summaries.
import matplotlib.pyplot as plt  # Import plotting tools for local regression curves.
from sklearn.linear_model import LinearRegression  # Import linear regression for a manual local weighted fit.
from statsmodels.nonparametric.smoothers_lowess import lowess  # Import statsmodels LOWESS smoother.
RANDOM_SEED = 123  # Store the reproducibility seed.
np.random.seed(RANDOM_SEED)  # Fix legacy NumPy randomness.
plt.style.use('default')  # Use a predictable plotting style.
````

````python
def make_sine_regression_data(n=120, noise=0.35, random_state=123):  # Define a one-dimensional nonlinear regression simulator.
    rng = np.random.default_rng(random_state)  # Create a reproducible random generator.
    x = np.sort(rng.uniform(0.0, 3.0, size=n))  # Draw sorted x values for a clean smoother plot.
    signal = 4.0 + np.sin(3.0 * x)  # Define the true smooth signal.
    y = signal + rng.normal(0.0, noise, size=n)  # Add independent Gaussian noise.
    return x, y, signal  # Return predictor, response, and true signal.

def local_linear_predict(x_train, y_train, x0, bandwidth):  # Define a manual local weighted regression prediction.
    x_train = np.asarray(x_train).ravel()  # Convert x values to a flat array.
    y_train = np.asarray(y_train).ravel()  # Convert y values to a flat array.
    weights = np.exp(-0.5 * ((x_train - x0) / bandwidth) ** 2)  # Give nearby x values larger weights.
    model = LinearRegression()  # Create a local linear regression model.
    model.fit(x_train.reshape(-1, 1), y_train, sample_weight=weights)  # Fit the local model using weights.
    prediction = model.predict(np.array([[x0]]))[0]  # Predict at the target point.
    return prediction  # Return the local linear prediction.
````

````python
x, y, true_signal = make_sine_regression_data(n=120, noise=0.35, random_state=RANDOM_SEED)  # Simulate a local-regression problem.
span_values = [0.15, 0.35, 0.65]  # Define LOWESS spans from flexible to smooth.
plt.figure(figsize=(6, 4))  # Create a figure for LOWESS span comparison.
plt.scatter(x, y, s=20, alpha=0.45, label='observed data')  # Plot observed data points.
for span in span_values:  # Draw a LOWESS curve for each span.
    smoothed = lowess(y, x, frac=span, return_sorted=True)  # Fit statsmodels LOWESS with this span.
    plt.plot(smoothed[:, 0], smoothed[:, 1], label=f'LOWESS frac={span}')  # Add the smooth curve to the plot.
plt.xlabel('x')  # Label the predictor axis.
plt.ylabel('y')  # Label the response axis.
plt.title('LOWESS span controls local flexibility')  # Title the span comparison plot.
plt.legend()  # Show span labels.
plt.show()  # Render the LOWESS plot.
manual_prediction = local_linear_predict(x, y, x0=1.5, bandwidth=0.40)  # Compute one manual local weighted regression prediction.
print(f'Manual local linear prediction at x=1.5: {manual_prediction:.3f}')  # Print the manual prediction.
````
