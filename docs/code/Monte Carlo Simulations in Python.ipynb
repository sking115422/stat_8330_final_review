## Simulation of Brownian Motion

Using the techniques that we have learned, I will demonstrate a quick simulation of Brownian motion. Brownian motion is a simple form of random movement. Albert Einstein first developed the mathematical theory to explain the motion of particles suspended in a liquid. The mathematical process is used as a model for many types of motion. I have used it in my research for modeling movement of animals called nematodes: 

We assume

1. The nematodes all start at a single point.   
2. The move in 2 dimensions.   
3. Time is divided into discrete increments. In each increment, the distance that a nematode moves is distributed as Normal(0,1) in the X-dimension and Normal(0,1)  in the Y-dimension.   
4. Movement is independent between time steps.   
5. Movement is independent between nematodes.   

The simulation code is below. The steps of the simulation:

1. Use `npr.normal` to generate random movements in the x direction for each nematode and time step. All of the x movements are generated at one time. There are `n` nematodes and `T` time steps. Thus, a total fo `n*T` movements are generated. They are all drawn from the same probability distribution.  
2. Arrange the `n*T` values into an array with `n` rows and `T` columns. Thus, each row is the movements of a single nematode for `T` time steps. 
3. Following exactly the same procedure as in steps 1 and 2, the random y movements are generated also.   
4. The starting positions are initialized to zero.  
5. I use the function `np.cumsum` (cumulative sum) to add up the random movements on x and in y to get the displacement in x and y.  
6. Plot the  x and y motion.  


```python
T=10 #number  of time steps
n=4 # number of nematodes

#generate the change in x position and change  in y position for each  nematode 
#and each time step:
xdelta=pd.DataFrame(npr.normal(size=T*n).reshape((n,T))) #change in x position
ydelta=pd.DataFrame(npr.normal(size=T*n).reshape((n,T))) #change in y position


#take starting position as x=0,y=0
xdelta.iloc[:,0]=0
ydelta.iloc[:,0]=0

#add up the changes in x and changes in y to get the displacement:
xdis=xdelta.apply(np.cumsum,1)
ydis=ydelta.apply(np.cumsum,1)



```

`xdelta` and `ydelta` contain the changes in x and position.  The  rows correspond to different nematodes and the columns correspond to different time steps. Each value is generated from a normal distribition with mean 0. 

```python
xdelta
```

The values in each row of `xdelta` are added up to get the current x position. The function `np.cumsum` adds up the elements of a vector and gives the sum at each step:

```python
np.cumsum([1,2,3,4,5])
```

Apply this to the rows of `xdelta`:

```python
xdis=xdelta.apply(np.cumsum,1)
xdis
```

xdis[x,y] is  the position of nematode x at time y. 

Next, I make a plot for a single nematode:

```python
plt.plot(xdis.iloc[0,:],ydis.iloc[0,:])
plt.xlabel("x position")
plt.ylabel("y position")
plt.title("Simulation of Brownian Motion for one nematode")
```

Now, I will plot the movment of 20 nematodes together. 

Note the use of the `for` loop to do the 20 plots.

```python
T=20 #number  of time steps
n=20# number of critters

#generate the change in x position and change  in y position for each  nematode 
#and each time step:
xdelta=pd.DataFrame(npr.normal(size=T*n).reshape((n,T))) #change in x position
ydelta=pd.DataFrame(npr.normal(size=T*n).reshape((n,T))) #change in y position

#Initialize the starting positon to (x,y)=(0,0)
xdelta.iloc[:,0]=0
ydelta.iloc[:,0]=0

#add up the changes in x and changes in y to get the displacement:
xdis=xdelta.apply(np.cumsum,1)
ydis=ydelta.apply(np.cumsum,1)

#Make the plot
plt.figure()
plt.xlabel("x position")
plt.ylabel("y position")
plt.title("Simulation of Brownian Motion with 20 critters")
for g in np.arange(n):  #loop over the nematodes and make a plot for each one
    plt.plot(xdis.iloc[g,:],ydis.iloc[g,:])

```
