## What problem are we solving?

Hierarchical clustering builds a tree of nested clusters by repeatedly merging nearby observations or clusters.

**Highest-probability exam pattern:** Compute a dendrogram, choose a number of clusters from the tree, and fit agglomerative clustering.

## Assumptions, intuition, and theory

Ward linkage merges clusters to control within-cluster variance. The dendrogram height shows merge distance. Like k-means, hierarchical clustering is exploratory and depends on scaling and distance choices.

## Python method notes and documentation

`linkage` computes the merge tree, `dendrogram` visualizes it, and `AgglomerativeClustering.fit_predict` returns cluster labels for a chosen number of clusters.

## Fully self-contained worked example

Every code line below is commented so it is easy to scan under exam time pressure.

````python
import numpy as np  # Import numerical arrays and random-number tools.
import pandas as pd  # Import DataFrame tables for cluster summaries.
import matplotlib.pyplot as plt  # Import plotting tools for dendrograms.
from scipy.cluster.hierarchy import dendrogram, linkage  # Import hierarchical linkage and dendrogram functions.
from sklearn.cluster import AgglomerativeClustering  # Import sklearn agglomerative clustering.
from sklearn.datasets import make_blobs  # Import cluster simulator.
from sklearn.metrics import adjusted_rand_score  # Import clustering recovery metric for simulated labels.
from sklearn.preprocessing import StandardScaler  # Import standardization before distance-based clustering.
RANDOM_SEED = 123  # Store the reproducibility seed.
np.random.seed(RANDOM_SEED)  # Fix legacy NumPy randomness.
plt.style.use('default')  # Use a predictable plotting style.
````

````python
def make_cluster_data(n=180, centers=3, cluster_std=0.8, random_state=123):  # Define a labeled cluster simulator for recovery checks.
    X, true_labels = make_blobs(  # Generate blob-shaped clusters with known labels.
        n_samples=n,  # Set the number of observations.
        centers=centers,  # Set the number of true clusters.
        cluster_std=cluster_std,  # Control cluster overlap.
        random_state=random_state,  # Fix the simulation seed.
    )  # Finish the simulator call.
    return X, true_labels  # Return data and true labels.
````

````python
X, true_labels = make_cluster_data(n=75, centers=3, cluster_std=0.8, random_state=RANDOM_SEED)  # Simulate a smaller data set for readable dendrograms.
X_scaled = StandardScaler().fit_transform(X)  # Standardize variables before distance-based clustering.
Z = linkage(X_scaled, method='ward')  # Compute Ward hierarchical linkage matrix.
model = AgglomerativeClustering(n_clusters=3, linkage='ward')  # Create a final agglomerative clustering model.
labels = model.fit_predict(X_scaled)  # Fit the model and return cluster labels.
ari = adjusted_rand_score(true_labels, labels)  # Compare recovered clusters to simulation truth.
print(f'Adjusted Rand index for hierarchical clustering: {ari:.3f}')  # Print recovery score.
plt.figure(figsize=(8, 4))  # Create a dendrogram figure.
dendrogram(Z, truncate_mode='level', p=4)  # Draw a truncated dendrogram for readability.
plt.xlabel('observation or merged cluster')  # Label dendrogram leaves.
plt.ylabel('linkage distance')  # Label merge height.
plt.title('Hierarchical clustering dendrogram')  # Title the dendrogram.
plt.show()  # Render the dendrogram.
plt.figure(figsize=(6, 4))  # Create a cluster scatterplot figure.
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, s=30, alpha=0.75)  # Plot observations colored by cluster assignment.
plt.xlabel('scaled x1')  # Label the first axis.
plt.ylabel('scaled x2')  # Label the second axis.
plt.title('Agglomerative clustering assignments')  # Title the scatterplot.
plt.show()  # Render the cluster plot.
````

## What to remember for the exam

A dendrogram is a clustering history. Cutting it lower gives more clusters; cutting it higher gives fewer clusters.
