## What problem are we solving?

PCA reduces many numeric variables to a smaller set of orthogonal directions that explain the most variance.

**Highest-probability exam pattern:** Standardize variables, fit PCA, report explained variance, draw a scree plot, and interpret the first two score directions.

## Assumptions, intuition, and theory

PCA is unsupervised, so it does not use response labels. Standardization is important when variables are on different scales. Components are linear combinations of original variables.

## Python method notes and documentation

`StandardScaler.fit_transform` standardizes features, `PCA.fit_transform` computes scores, `explained_variance_ratio_` reports variance fractions, and score plots show low-dimensional structure.

## Fully self-contained worked example

Every code line below is commented so it is easy to scan under exam time pressure.

````python
import numpy as np  # Import numerical arrays and random-number tools.
import pandas as pd  # Import DataFrame tables for explained variance summaries.
import matplotlib.pyplot as plt  # Import plotting tools for scree and score plots.
from sklearn.datasets import make_blobs  # Import cluster simulator for visible PCA groups.
from sklearn.decomposition import PCA  # Import principal component analysis.
from sklearn.preprocessing import StandardScaler  # Import standardization before PCA.
RANDOM_SEED = 123  # Store the reproducibility seed.
np.random.seed(RANDOM_SEED)  # Fix legacy NumPy randomness.
plt.style.use('default')  # Use a predictable plotting style.
````

````python
def make_cluster_data(n=180, centers=3, cluster_std=0.8, random_state=123):  # Define a labeled cluster simulator for recovery checks.
    X, true_labels = make_blobs(  # Generate blob-shaped clusters with known labels.
        n_samples=n,  # Set the number of observations.
        centers=centers,  # Set the number of true clusters.
        cluster_std=cluster_std,  # Control cluster overlap.
        random_state=random_state,  # Fix the simulation seed.
    )  # Finish the simulator call.
    return X, true_labels  # Return data and true labels.
````

````python
X, true_labels = make_cluster_data(n=180, centers=3, cluster_std=1.0, random_state=RANDOM_SEED)  # Simulate multivariate data with known groups.
X_extra = np.column_stack([X, 0.8 * X[:, 0] - 0.3 * X[:, 1]])  # Add a correlated third variable to make PCA meaningful.
scaler = StandardScaler()  # Create a standardizer so variables contribute comparably.
X_scaled = scaler.fit_transform(X_extra)  # Standardize variables before PCA.
pca = PCA(n_components=3)  # Create PCA keeping all three components.
scores = pca.fit_transform(X_scaled)  # Fit PCA and compute component scores.
variance_table = pd.DataFrame({'component': [1, 2, 3], 'explained_variance_ratio': pca.explained_variance_ratio_})  # Build explained variance table.
display(variance_table)  # Display the variance table.
plt.figure(figsize=(6, 4))  # Create a scree-plot figure.
plt.plot(variance_table['component'], variance_table['explained_variance_ratio'], marker='o')  # Plot explained variance by component.
plt.xlabel('principal component')  # Label the component axis.
plt.ylabel('explained variance ratio')  # Label the variance axis.
plt.title('PCA scree plot')  # Title the scree plot.
plt.show()  # Render the scree plot.
plt.figure(figsize=(6, 4))  # Create a score-plot figure.
plt.scatter(scores[:, 0], scores[:, 1], c=true_labels, s=30, alpha=0.75)  # Plot the first two component scores colored by true group.
plt.xlabel('PC1 score')  # Label the first component axis.
plt.ylabel('PC2 score')  # Label the second component axis.
plt.title('PCA score plot')  # Title the score plot.
plt.show()  # Render the score plot.
````

## What to remember for the exam

PCA does not know the response. If class labels separate in a PCA score plot, that is a discovered structure, not something PCA was trained to do.
